```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# import an example dataset from the seaborn library
tips = sns.load_dataset('tips')
tips.head()
```

```python
# distribuition plot of one variable, histogram with lineplot with the KDE (kernel density estimation)
# really cool!
sns.distplot(tips['total_bill'], bins=20)
```

```python
# joint plot, combine two variables distribuition
# for instance, lets compare the distrbuition of the total bill with the tip value
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
# the scatter plot shows the trend of relation between the two distribuitions
# we can use the kind argument, specifiing 'hex' instead of the default scatter plot
# the hex polygons gets darker if there are more points falling inside of it
# or we can use the 'reg' to draw a regression line that fits the plot
# or we can use the 'kde' to draw a two dimension kde that represents the density of points
```

```python
# extending the concept of jointplot we have pairplot, that essentially does the samething but for
# all columns in the dataframe, instead of only two
sns.pairplot(tips, hue='sex', palette='coolwarm')
# it gives a scatterplot comparing two different columns, and a histogram when comparing
# the same variable
# we can add a hue argument to add columns that are categorical, instead of numerical
# in our dataset the sex column is categorical, Male or Female, so it would color the data points
# associated with each category
# also, we can add a palette argument to use a specific color for our plots
```

```python
# rugplot help us understand the concept of KDE
# it just draws a line for every single point along the variable axis
# The KDE is contructed summing the gaussian distribuition curves generated for each line, drawed by the rugplot
sns.rugplot(tips['total_bill'])
```

```python
# we can have only the KDE plot alone:
sns.kdeplot(tips['total_bill'])
```
